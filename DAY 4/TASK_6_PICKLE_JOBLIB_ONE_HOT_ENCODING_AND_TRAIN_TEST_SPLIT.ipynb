## Dataset Loading and Serialization Demonstration

```python
import pandas as pd
import pickle
import joblib
import os
```

### 1. Load the dataset

```python
file_path = '/content/ai_product_launch_dataset.csv.csv'
df = pd.read_csv(file_path)
display(df.head())
```

### 2. Demonstrate `pickle` for saving and loading the DataFrame

```python
pickle_filename = 'dataframe.pkl'

with open(pickle_filename, 'wb') as f:
    pickle.dump(df, f)
print(f"DataFrame successfully pickled and saved to {pickle_filename}")

with open(pickle_filename, 'rb') as f:
    loaded_df_pickle = pickle.load(f)
print(f"DataFrame successfully loaded from {pickle_filename}")
display(loaded_df_pickle.head())
```

### 3. Demonstrate `joblib` for saving and loading the DataFrame

```python
joblib_filename = 'dataframe.joblib'

joblib.dump(df, joblib_filename)
print(f"DataFrame successfully saved using joblib to {joblib_filename}")

loaded_df_joblib = joblib.load(joblib_filename)
print(f"DataFrame successfully loaded from {joblib_filename}")
display(loaded_df_joblib.head())
```

### 4. Perform One-Hot Encoding on Categorical Features

```python
categorical_cols = df.select_dtypes(include=['object', 'bool']).columns

categorical_cols = [col for col in categorical_cols if col not in ['crawl_timestamp', 'product_url', 'product_name', 'description', 'image', 'product_category_tree', 'product_specifications']]

print(f"Categorical columns identified for one-hot encoding: {categorical_cols}")

df_encoded = pd.get_dummies(df, columns=categorical_cols, drop_first=True)

print("\nDataFrame after one-hot encoding:")
display(df_encoded.head())
```

4.PERFORM TRAIN_TEST-SPLIT OPERATION

```python
from sklearn.model_selection import train_test_split

X = df_encoded.drop('discounted_price', axis=1)
y = df_encoded['discounted_price']

X = X.select_dtypes(include=['number', 'bool'])

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

print(f"Original features shape: {X.shape}")
print(f"Training features shape: {X_train.shape}")
print(f"Testing features shape: {X_test.shape}")
print(f"Training target shape: {y_train.shape}")
print(f"Testing target shape: {y_test.shape}")
```
